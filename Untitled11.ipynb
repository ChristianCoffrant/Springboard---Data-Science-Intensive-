We can see that some of the data types are of type object, since they are categorical in nature it will be useful for us to convert these to the 'category' data type

```python
for col in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome', 'y']:
    bank[col] = bank[col].astype('category')

# Print off the data types to check that our for loop worked as intended
bank.dtypes
```

We will use the Freedmanâ€“Diaconis rule to estimate the ammounts of bins we should use for our visualisations

```python
def IQR(dist):
    return np.percentile(dist, 75) - np.percentile(dist, 25)

def binSize(col):
    return 2*((IQR(bank[col]))/(bank[col].count()**(1/3)))

#print(bank['age'].count())
#print(binSize(bank['age']))  

       
for col in ['age', 'balance', 'day', 'duration', 'campaign', 'pdays', 'previous']:
    #print(binSize(col))
    print("Reccomended bin size for "+str(col)+":\t"+ str(bank[col].max()-bank[col].min()/binSize(col)))
```

Now that we have some idea of the shape and distribution of the data, let's train a decision tree. We start by importing the required libaries

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

We then split the data into a training and testing dataframe. Note that we drop the y column which is our outcome column i.e. the variable that we're trying to predict. However aso notice that we drop the 'poutcome' column, this column was the banks attempt at predicting the outcome, we don't want to train our model using their model, however it may be interesting to compare their predictions to those of the model which we build

```python
X = bank.drop(['y', 'poutcome'], axis=1)

# The decision tree function doesn't like categorical data, we can get around this by using category codes inplace of categories 
for col in ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month']:
    X[col] = X[col].cat.codes
    
```

We train the model on 70% of the data and use the other 30% to evaluate the accuracy of our model.

```python
Y = bank.y
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.3)

dTree= DecisionTreeClassifier()
dTree.fit(X_train,Y_train)
predictions = dTree.predict(X_test)
```

```python
print(classification_report(Y_test,predictions))
```

The model yields an 87% accuracy, that is quite good! However let's see if we can't improve that by training a random forrest model

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, Y_train)
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(Y_test,rfc_pred))
```

Although our random forest model only yields a 2% increase in accuracy, there's a big improvement in it's ability to predict accuratly whether or not someone will signup for a term deposit, we can see this in the increase of the yes precision, which has increased from 44% to 60%. This precision will serve as a nice baseline for evaluating other machine learning models
